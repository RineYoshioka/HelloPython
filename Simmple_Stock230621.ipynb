```python
# JRE,JRW,JRC Stock data from Stooq.com
# reference https://seanmemo.com/234/

!pip install mplfinance
import datetime
from pandas_datareader import data
dt2 = datetime.datetime.now()
dt1 = dt2 - datetime.timedelta(weeks=2)
start = dt1.strftime('%Y-%m-%d')
end = dt2.strftime('%Y-%m-%d')
df1 = data.DataReader('9020.JP', 'stooq', start, end)

```

```python
import mplfinance as mpf
mpf.plot(df1, type='candle', figratio=(12,4))
```

```python
c2='9022.jp'
df2=data.DataReader(c2, 'stooq', start, end)
mpf.plot(df2, type='candle', volume=True, figratio=(12,4))
```

```python
dt2 = datetime.datetime.now()
dt1 = dt2 - datetime.timedelta(weeks=4)
start = dt1.strftime('%Y-%m-%d')
end = dt2.strftime('%Y-%m-%d')
c3 = '9021.JP'
df3 = data.DataReader(c3, 'stooq', start, end)
mpf.plot(df3, type='candle', volume=True, mav=5, figratio=(12,4))
```
